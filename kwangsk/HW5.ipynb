```python
from scipy import *
from scipy.special import jn, jn_zeros
def drumhead_height(n,k,r,theta,t):
    return cos(t) * cos(n*theta) * jn(n, r*(jn_zeros(n,k)[k-1]))
```

```python
from mpl_toolkits.mplot3d import axes3d
def plot_Bessel (t):
    r = linspace(0,1)
    theta = linspace(0, 2*pi)
    r_mesh, theta_mesh = meshgrid(r,theta)
    x = r_mesh*cos(theta_mesh)
    y = r_mesh*sin(theta_mesh)
    fig = plt.figure(figsize=(16,6))
    gs = plt.GridSpec(3, 3, wspace=0.2, hspace=0.2) 
    def plot_subplot(s, n, k):
        ax= fig.add_subplot(s, projection='3d')
        ans = drumhead_height(n,k,r_mesh,theta_mesh,t)
        style = {
                'vmin': -0.5,
                'vmax': 0.5,
                'cmap': cm.jet,
                'alpha': 0.4,
                }
        surf = ax.plot_surface(x, y, ans, rstride=1, cstride=1, **style)
        ax.contourf(x, y, ans, zdir='z', offset=-0.5, **style)
        ax.contour(x, y, ans, zdir='y', offset=1, **style)
        ax.contour(x, y, ans, zdir='x', offset=-1, **style)
        fig.colorbar(surf)
        
        # set up title and labels
        ax.set_title('$u_{{{},{}}}$'.format(n, k))
        ax.set_xlabel('X')
        ax.set_xlim(-1, 1)
        ax.set_ylabel('Y')
        ax.set_ylim(-1, 1)
        ax.set_zlabel('Z')
        ax.set_zlim(-0.5, 0.5)

    # lay out multiple subplots
    plot_subplot(gs[:3,:2], 2, 3)
    plot_subplot(gs[0,2], 4, 1)
    plot_subplot(gs[1,2], 1, 4)
    plot_subplot(gs[2,2], 3, 3)
    
```

```python
plot_Bessel(0)
plt.savefig('0.png')
plot_Bessel(1)
plt.savefig('1.png')
plot_Bessel(2)
plt.savefig('2.png')
plot_Bessel(3)
plt.savefig('3.png')
plot_Bessel(4)
plt.savefig('4.png')
```
